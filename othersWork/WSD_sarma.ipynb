```python
import math

 

def getInput(data, testingData):

    noOfEntries = int(input("Enter no. of entries:"))

 

    for i in range(noOfEntries):

        wordList = []

        lengthOfEntry = int(input("Enter No. of Words in the Entry: "))

        for j in range(lengthOfEntry):

            wordList.append(input(f"word{j+1}:"))

        cls = input("Enter class for this entry: ")

        entry = (wordList, cls)

        data.append(entry)

 

    print("Now enter testing data:")

    lengthOfTesting = int(input("Enter No. of Words : "))

    for i in range(lengthOfTesting):

        testingData.append(input(f"word{i+1}:"))

 

 

def countClasses(data, classList):

    classCountDict = {}

    for i in range(len(data)):

        classCountDict[data[i][1]] = 0

        if data[i][1] not in classList:

            classList.append(data[i][1])

    for i in range(len(data)):

        classCountDict[data[i][1]] += 1

    return classCountDict

 

 

def calculateClassProbabilities(classCountDict):

    totalClass = sum(classCountDict.values())

    classProbDict = {}

    for key in classCountDict:

        classProbDict[key] = classCountDict[key] / totalClass

    return classProbDict

 

 

def getTotalVocabulary(data):

    vocabCount = 0

    wordList = []

    for entry in data:

        for word in entry[0]:

            if word not in wordList:

                vocabCount += 1

                wordList.append(word)

 

    return vocabCount

 

 

def getCountWordWithClass(data, word, cls):

    count = 0

    for entry in data:

        if (entry[1] == cls) and (word in entry[0]):

            count += 1

    return count

 

 

def bayes(testingData, data, classCountDict, classProbDict, classList):

    maxProbClass = 0

    optClass = 'f'

    for cls in classList:

        prod = classProbDict[cls]

        for word in testingData:

            ent = (getCountWordWithClass(data, word, cls)) / classCountDict[cls]

            prod *= ent

        if prod > maxProbClass:

            maxProbClass = prod

            optClass = cls

    return optClass

 

 

def smoothing(testingData, data, classCountDict, classProbDict, classList, vocabCount):
    maxProbClass = 0
    optClass = 'f'

    for cls in classList:

        prod = classProbDict[cls]

        for word in testingData:

            ent = (getCountWordWithClass(data, word, cls) + 1) / (classCountDict[cls] + vocabCount)

            prod *= ent

        if prod > maxProbClass:

            maxProbClass = prod

            optClass = cls

    return optClass

 

 

def smoothingScore(testingData, data, classCountDict, classProbDict, classList, vocabCount):

    maxProbClass = 0

    optClass = 'f'

    for cls in classList:

        prod = classProbDict[cls]

        for word in testingData:

            ent = (getCountWordWithClass(data, word, cls) + 1) / (classCountDict[cls] + vocabCount)

            prod *= ent

        if prod > maxProbClass:

            maxProbClass = prod

            optClass = cls

    return math.log(maxProbClass)

 

 

data = []  

testingData = []  

getInput(data, testingData)

print(data)

print(testingData)

classList = []

classCountDict = countClasses(data, classList)

print(classCountDict)

print(classList)

classProbDict = calculateClassProbabilities(classCountDict)

print(classProbDict)

vocabCount = getTotalVocabulary(data)

print(vocabCount)

classPredictBayes = bayes(testingData, data, classCountDict, classProbDict, classList)

print(f"With Bayes class: {classPredictBayes}")

classPredictSmoothing = smoothing(testingData, data, classCountDict, classProbDict, classList,

                                  vocabCount)

print(f"With Smoothing class: {classPredictSmoothing}")

classPredictScore = smoothingScore(testingData, data, classCountDict, classProbDict, classList,

                                   vocabCount)

print(f"With score class: {classPredictSmoothing}, Score: {classPredictScore}")
```
